```python
import torch
import random
import matplotlib.pyplot as plt
```

```python
def get_test(times, nums):
    total=torch.zeros(nums)
    P=torch.zeros(times)
    x=torch.arange(times)
    for num in range(times):
        tmp=random.randint(0,nums-1)
        total[tmp]+=1
        P[num]=total[0]/(num+1)
    ans=torch.ones(times)/(nums)
    fig, ax = plt.subplots()
    ax.plot(x, P, label='P')
    ax.plot(x, ans, label="ans")
    ax.legend()

    
    
```

```python
get_test(100000, 10)
```

```python
get_test(100000, 2)
```
