```python
import matplotlib.pyplot as plt
import torch
import math
```

```python
def get_grad(f, x, delta):
    df=f(x+delta)-f(x)
    return df/(delta*1.)
```

```python
def sin(x):
    return math.sin(x)
```

```python
delta=0.1
for i in range(10):
    x=get_grad(sin, 0, delta)
    delta*=0.1
    print(f"grad={x:.10f}")
```

```python
def get_grad_tensor(f, x, delta):
    df=f(x+delta)-f(x)
    return df/(delta*1.)
def sin_tensor(x):
    return torch.sin(x)
```

```python
x=torch.arange(0,10,0.1)

y=get_grad_tensor(sin_tensor, x, 0.001)
```

```python
fig, ax = plt.subplots()# 图实例
ax.plot(x, y, label="sin'(x)")
ax.plot(x, torch.cos(x), label="cos(x)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend() #自动检测要在图例中显示的元素，并且显示
fig1, ax1 = plt.subplots()  #另一个实例
plt.show()
```
